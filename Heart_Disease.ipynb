Multivariate analysis with HeartDisease

```python
%matplotlib inline
for i in range(len(data.columns)):
    if is_numeric_dtype(data.iloc[:,i])==True:
        y = data.iloc[:,i]
        nom = data.columns[i]
        plt.figure()
        print(sns.boxplot(y=y,x=data.loc[:,"HeartDisease"]))
        #plt.show()
    else:
        x=data.iloc[:,i]
        plt.figure()
        print(sns.countplot(x=x,hue=data.loc[:,"HeartDisease"]))
        #plt.show()
#plt.show()
```

For different categorical data we are going to analyze by percentage of each HeartDisease group.
The variables we are going to see are the categorical ones

```python
%matplotlib inline
for i in range(len(data.columns)):
    if is_numeric_dtype(data.iloc[:,i])==False and i!=0:
        x='HeartDisease'
        y = data.columns[i]
        plt.figure()
        print(data.groupby(x)[y].value_counts(normalize=True).mul(100).rename('percent').reset_index().pipe((sns.catplot,'data'), x=x,y='percent',hue=y,kind='bar'))
        #plt.show()
```

In a first look we can take a first impression:
- The people who smokes has a higher percentage of haveing a HeartDisease* 
- The people affected by a Stroke are more likely to a HeartDisease or viceversa*
- People with higher rate of DiffWalking are more likely to a HeartDisease*
- There are more men than women with HeartDisease
- The GenHealth is more extreme in the HeartDisease population
- In the KidneyDisease the number of HeartDisease are higher
- SkinCancer as in KidneyDisease the number of HeartDisease are higher
